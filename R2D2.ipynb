```python
import urllib.request
from bs4 import BeautifulSoup
```

```python
url = urllib.request.urlopen("https://www.imdb.com/search/title/?count=100&groups=top_1000&sort=user_rating")
ht = url.read()
so = BeautifulSoup(ht,"html.parser")
```

```python
tit=so.title
print(tit)
```

```python
name=[]
rating=[]
cast=[]
id=[]

pgs=so.select("div.lister-item-content")
for i in pgs:
    name.append((i.find("a").text))
    
    rating.append(((i.find('div','inline-block ratings-imdb-rating')).text).split()[0])
    
    id.append(((i.find("a"))['href']).split('/')[2])
    
    list = i.findAll('p')
    list = list[2].text
    ind = list.find("Stars:")
    newl = ("".join(list[ind+6:].split('\n'))).strip()
    newl = newl.split(', ')
    cast.append(newl)
    
```
